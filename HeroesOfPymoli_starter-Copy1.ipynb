## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group the original DataFrame by Name
top_spenders = purchase_data.groupby("SN")

# Create a Series with "SN" as the index and purchase count as the value
spender_purchase_count = top_spenders["Price"].count()

# Find the total amount spent for each Name
spender_total = top_spenders["Price"].sum()

# Calculate the average price spent for each screenname by dividing their total amount spent by the number of purchases
spender_avg_price = spender_total/spender_purchase_count

# Create a DataFrame with calculated values
spender_df = pd.DataFrame({"Purchase Count": spender_purchase_count,
                          "Average Purchase Price": spender_avg_price,
                          "Total Purchase Value": spender_total})

# Sort DataFrame by "Total Purchase Value" column in descending order
spender_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

# Convert "Average Purchase Price" and "Total Purchase Value" columns to curry format
spender_df["Average Purchase Price"] = spender_df["Average Purchase Price"].map("${:,.2f}".format)
spender_df["Total Purchase Value"] = spender_df["Total Purchase Value"].map("${:,.2f}".format)

# Display first 5 rows of DataFrame
spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DataFrame from a subsection of original DataFrame
item_df = purchase_data.loc[:, ['Item ID', 'Item Name', 'Price']]

# Create a groupby object by grouping the item DataFrame by "Item ID" and "Item Name" columns
item_group = item_df.groupby(["Item ID", "Item Name"])

# Find the number of times the item was purchased
item_count = item_group['Price'].count()

# Find the total amount spent on the item
item_total = item_group['Price'].sum()

# Find the price of the individual item by dividing the total amount spent by the purchase count
item_price = item_total/item_count



# Create a DataFrame displaying the above values
item_summary = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_price,
                             "Total Purchase Value": item_total})

# Sort the DataFrame by "Purchase Count" column in descending order 
item_sorted = item_summary.sort_values(by=['Purchase Count'], ascending=False)

# Convert the "Item Price" and "Total Purchase Value" columns to currency format
item_sorted["Item Price"] = item_sorted["Item Price"].map("${:,.2f}".format)
item_sorted["Total Purchase Value"] = item_sorted["Total Purchase Value"].map("${:,.2f}".format)

# Display the first 5 rows of the DataFrame
item_sorted.head()
```
